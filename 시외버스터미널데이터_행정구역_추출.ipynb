```python
import requests

def reverse_geocode(lat, lng, api_key):
    # Google Maps Geocoding API 엔드포인트
    endpoint = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lng}&key={api_key}&language=ko"

    # API 호출
    response = requests.get(endpoint)
    if response.status_code == 200:
        results = response.json()
        if results['status'] == 'OK':

            # 첫 번째 결과의 주소 반환
            #return dict(results)
            #return results['results'][0]['formatted_address']
            #return results['results'][0]['plus_code']['compound_code'][7:].replace(' ','')
            return results
        else:
            return "Geocoding API 요청 실패: " + results['status']
    else:
        return "Geocoding API 요청 실패"

# 테스트용 위도와 경도
latitude = 37.5665
longitude = 126.9780

# Google Cloud Console에서 발급받은 API 키
api_key = 'AIzaSyDeWn3K49QFqYcsfc_ZPTDG7Va1hX2TumU'

# 좌표를 주소로 변환
address = reverse_geocode(latitude, longitude, api_key)
print("위도:", latitude)
print("경도:", longitude)
print("주소:", address)
```
