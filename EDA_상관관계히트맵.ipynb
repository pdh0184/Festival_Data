```python
data = data.fillna(0)
data
```

```python
data['예산합계'] =data['예산합계'].apply(lambda row:float(row.replace(',', '')))
data['방문객 합계'] = data['방문객 합계'].str.replace(',','')
data['방문객 합계'] = pd.to_numeric(data['방문객 합계'])
```

# 이상치 제거

```python
df_non_outer = data.copy()
cols = ['공휴일수', '축제 기간', '방문객 합계']

outlier_high = []
for col in cols:
    if col == '공휴일수':
        outlier_high.append(15)
        print(f"{col}의 높은 이상치 기준 : {15.0}")
        continue
    else:
        col_f1 = data.describe()[col]['25%']
        col_f3 = data.describe()[col]['75%']

        col_IQR = (col_f3 - col_f1)
        col_outlier_high = col_f3 + 1.5 * col_IQR * 10
        outlier_high.append(col_outlier_high)
        print(f"{col}의 높은 이상치 기준 : {col_outlier_high:.3f}")

for i in range(len(cols)):
    df_non_outer = df_non_outer[df_non_outer[cols[i]] <= outlier_high[i]]

data = df_non_outer
data
```

#sweetbiz를 사용하여 빈도수 히스토그램 확인
